```python
import pandas as pd

# Read the CSV files into DataFrames
ticketmaster_df = pd.read_csv('data/ticketmaster_data.csv')
spotify_df = pd.read_csv('data/spotify_data.csv')

# Merge the two DataFrames on the 'artist' column
merged_df = pd.merge(ticketmaster_df, spotify_df, on='artist', how='inner')  # 'inner' for intersection, 'outer' for union

# Display the merged DataFrame
print(merged_df)

# Optionally, save the merged DataFrame to a new CSV
merged_df.to_csv('data/merged_artist_data.csv', index=False)
```
